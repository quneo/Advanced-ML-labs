```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report



data = pd.read_csv("data/ethnic female.csv", sep=";")
print(data.head())
print(f"\nВсего записей: {len(data)}")
print(data['национальность'].value_counts())
```

```python
required_columns = ["рост", "масса", "ширина плеч", "ширина таза"]

for col in required_columns:
    data[col] = data[col].astype(str).str.replace(",", ".", regex=False)


for col in required_columns:
    data[col] = data[col].astype(float)
```

```python
russian = data[data['национальность'] == 'рус'].head(300)
khakass = data[data['национальность'] == 'хак'].head(300)

df_balanced = pd.concat([russian, khakass], ignore_index=True)
print(f"Отобрано: русских - {len(russian)}, хакасок - {len(khakass)}")
print(f"Всего: {len(df_balanced)}")
```

```python
X = df_balanced[['рост', 'масса', 'ширина плеч', 'ширина таза']]
y = df_balanced['национальность']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Обучающая выборка: {len(X_train)}")
print(f"Контрольная выборка: {len(X_test)}")
print(f"\nРаспределение в обучающей:")
print(y_train.value_counts())
print(f"\nРаспределение в контрольной:")
print(y_test.value_counts())
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_pred = gnb.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Точность классификации (гауссовское распределение): {accuracy:.4f}")
print(f"\nМатрица ошибок:")
print(confusion_matrix(y_test, y_pred))
print(f"\nОтчет по классам:")
print(classification_report(y_test, y_pred))
```

```python
errors = (y_pred != y_test).sum()
error_rate = 1 - accuracy

print(f"Количество ошибок: {errors} из {len(y_test)}")
print(f"Ошибка классификации: {error_rate:.4f} ({error_rate*100:.2f}%)")
print(f"Правильно классифицировано: {accuracy:.4f} ({accuracy*100:.2f}%)")
```
